```python
%%writefile domain.yml
intents:
  - greet
  - thanks
  - Business
  - Economy
  - Normal
  - Standing
  - t&d
  - rd
  - bye
  - search_flights
  - search_trains
  - find_itineraries

entities:
  - source
  - destination
  - date
  - time

responses:
  utter_rd:
  - text:"ok sir,do you want a round ticket?"
  utter_t&d:
  - text:"can you please tell me your preffered time and day"
  utter_search_flights:
  - text:"Yes.For sure.which class do you prefer? --Business class  -- Economy class"
  utter_search_trains:
  - text:"Yes.For sure.which type of ticket do you prefer? --Normal  -- Standing "
  utter_greet:
  - text: "Hey! How are you?"
  utter_booked:
  - text: "just a minute....your ticket has been booked!"
  utter_thanks:
  - text: "you are welcome!"
  utter_goodbye:
  - text: "Have a nice day!"


session_config:
  session_expiration_time: 60
  carry_over_slots_to_new_session: true
```

```python
%%writefile data/stories.md
## booking Business
 * greet
   - utter_greet
 * search_flights
   - utter_t&d
 * t&d
   - utter_rd
 * rd
   - utter_search_flights
 *Business
   - utter_booked
 * thanks
   - utter_thanks
 * bye
    - utter_goodbye

## booking Economy
 * greet
   - utter_greet
 * search_flights
   - utter_t&d
 * t&d
   - utter_search_flights
 *Economy
   - utter_booked
 * thanks
   - utter_thanks
 * bye
    - utter_goodbye

## booking Normal
 * greet
   - utter_greet
 * search_flights
   - utter_t&d
 * t&d
   - utter_search_flights
 * Normal
   - utter_booked
 * thanks
   - utter_thanks
 * bye
   - utter_goodbye

## booking Standing
 * greet
   - utter_greet
 * search_flights
   - utter_t&d
 * t&d
   - utter_search_flights
 * Standing
   - utter_booked
 * thanks
   - utter_thanks
 * bye
   - utter_goodbye
  
```
